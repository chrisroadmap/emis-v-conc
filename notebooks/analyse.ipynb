```python
import os

import numpy as np
import matplotlib.pyplot as pl
import pandas as pd
import xarray as xr
```

```python
os.makedirs('../plots', exist_ok=True)
```

```python
pl.style.use('../defaults.mplstyle')
```

```python
conc = xr.open_dataset('../results/conc.nc')
```

```python
emis = xr.open_dataset('../results/emis.nc')
```

```python
pl.fill_between(
    emis.timebound,
    np.percentile(emis.temperature[:, 0, :], 5, axis=1),
    np.percentile(emis.temperature[:, 0, :], 95, axis=1),
    color='k',
    lw=0,
    alpha=0.2
)
pl.plot(
    emis.timebound,
    np.median(emis.temperature[:, 0, :], axis=1),
    color='k'
)
pl.fill_between(
    conc.timebound,
    np.percentile(conc.temperature[:, 0, :], 5, axis=1),
    np.percentile(conc.temperature[:, 0, :], 95, axis=1),
    color='r',
    lw=0,
    alpha=0.2
)
pl.plot(
    conc.timebound,
    np.median(conc.temperature[:, 0, :], axis=1),
    color='r'
)
```
